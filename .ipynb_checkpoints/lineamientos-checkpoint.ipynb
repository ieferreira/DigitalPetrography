```python
import cv2
import numpy as np 
imagenCe = loadGrey("ChitagaCerrito.jpg")
imagen = canny(imagenCe, 100,200)

def findLines(bordes, rho, theta, thr, mll, mlg):
    lineas = cv2.HoughLinesP(bordes, rho, theta, thr, minLineLength=mll, maxLineGap=mlg)
    return lineas

def drawLines(lineas, imagen):  
    for linea in lineas:
        x1,y1,x2,y2 = linea[0]
        cv2.line(imagen, (x1,y1), (x2,y2), (255, 0,0), 1)
    return imagen

bordes = canny(imagen, 40,160)
@wd.interact(rho=(0.1,10.1,1),thr=(20,60,1), mll=(0,40,5), mlg=(0,120,5))


def actualice2(rho=1, thr=20, mll=10, mlg=15):
    #borde = canny(imagen, 70,160)
    imag = imgColor.copy()
    lineas = drawLines(findLines(bordes, rho,np.pi/180, thr, mll, mlg), imag)
    return graph(lineas)
```

```python
fname2 = "olivino.jpg"
imgOlivino = loadGrey(fname2)
imgOlivinoColor= load(fname2)
bordesOlivino = canny(imgOlivino, 80,160)

@wd.interact(rho=(0,10,1),thr=(20,60,1), mll=(0,40,5), mlg=(0,120,5))


def actualice2(rho=1, thr=30, mll=0, mlg=10):
    #borde = canny(imagen, 70,160)
    imag = imgOlivinoColor.copy()
    lineas = drawLines(findLines(bordesOlivino, rho,np.pi/180, thr, mll, mlg), imag)
    return graph(lineas)
```
