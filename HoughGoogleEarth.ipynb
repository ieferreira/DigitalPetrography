```python

%matplotlib inline
img = cv2.imread("ChitagaCerrito.jpg", cv2.IMREAD_COLOR)
gris = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) # <- bgr es como trata opencv

plt.imshow(gris, cmap="Greys")
```

```python
bordes = cv2.Canny(gris, 5,30)
cv2.imshow("img", bordes)

plt.imshow(bordes)
```
