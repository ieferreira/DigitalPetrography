```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

from sklearn.datasets import make_blobs, make_circles, make_moons
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn import metrics

from pylab import rcParams

rcParams['figure.figsize'] = 10, 10

```

```python
x, y = make_blobs(n_samples = 500, n_features = 2, centers = 3, cluster_std=1.9, shuffle=False, random_state=42)

df = pd.DataFrame(x)
df.columns = ["x", "y"]
```

```python
sns.lmplot("x", "y", data = df, fit_reg=False, height=7)
```

```python
kmeans = KMeans(n_clusters=3, init="k-means++")
```

```python
x[:5]
```

```python
kmeans.fit(x)
```

```python
print(df)
```

```python
df["clusters"] = kmeans.labels_
print(df)
```

```python
sns.lmplot("x","y", data = df, fit_reg =False, hue = "clusters", height = 7)
```

```python
metrics.completeness_score(labels_pred=df["clusters"],
                          labels_true=y) 
```

```python
kmeans = KMeans(n_clusters=8)
kmeans.fit(x)
df["clusters"] = kmeans.labels_
```

```python
sns.lmplot("x","y", data = df, fit_reg =False, hue = "clusters", height = 7)
```

```python
metrics.completeness_score(labels_pred=df["clusters"],
                          labels_true=y) 
```
